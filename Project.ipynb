```python
#Filtrando las columnas numericas

numeric_colums = data.select_dtypes(include='number').columns
df_numeric = data[numeric_colums]
df_numeric
```

```python
#El promedio por año de todos los datos exceptuando "artists"

df_result = df_numeric.groupby('year').mean().reset_index()
df_result
```
